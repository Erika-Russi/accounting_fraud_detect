```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
df = pd.read_csv('04.c.KL_divergence_with_Co_names.csv')
df.head()
```

```python
df.columns
```

```python
columns_to_avg = ['freq_1', 'freq_2','freq_3', 'freq_4', 'freq_5', 'freq_6', 'freq_7', 'freq_8', 'freq_9']
```

```python
Benford_frequencies = pd.DataFrame([])
for col in columns_to_avg:
    col = df.groupby(['company_name'])[col].agg(lambda x: x.unique().mean())
    Benford_frequencies = Benford_frequencies.append(col)
```
